# Simple text map generator
A set of scripts that will print a very simple text map.
```
#: water tiles
^: mountain tiles
": grass tiles
_: void tiles
```

### Imports

````python
import random
````

### Configuration
Adjust as you please.

````python
grid_size = 20
num_of_rivers = 2
num_of_mountain_clusters = 5
````

### Map generation

#### Map grid initialization
Initialize the map with void.

````python
map_grid = [['_' for _ in range(grid_size)] for _ in range(grid_size)]
````

#### River generation
Generate `num_of_rivers` rivers that cross the map.

````python
for _ in range(num_of_rivers):
    start_edge = random.choice(['top', 'bottom', 'left', 'right'])
    if start_edge == 'top':
        x, y = random.randint(0, grid_size-1), 0
    elif start_edge == 'bottom':
        x, y = random.randint(0, grid_size-1), grid_size-1
    elif start_edge == 'left':
        x, y = 0, random.randint(0, grid_size-1)
    else:
        x, y = grid_size-1, random.randint(0, grid_size-1)

    while 0 <= x < grid_size and 0 <= y < grid_size:
        map_grid[y][x] = '#'
        direction = random.choice(['up', 'down', 'left', 'right'])
        if direction == 'up':
            y -= 1
        elif direction == 'down':
            y += 1
        elif direction == 'left':
            x -= 1
        else:
            x += 1
````

#### Mountain generation
Generate `num_of_mountain_clusters` mountain clusters.

````python
for _ in range(num_of_mountain_clusters):
        x, y = random.randint(0, grid_size-1), random.randint(0, grid_size-1)
        cluster_size = random.randint(5, 15)
        for _ in range(cluster_size):
            if 0 <= x < grid_size and 0 <= y < grid_size and map_grid[y][x] == '_':
                map_grid[y][x] = '^'
            direction = random.choice(['up', 'down', 'left', 'right'])
            if direction == 'up':
                y -= 1
            elif direction == 'down':
                y += 1
            elif direction == 'left':
                x -= 1
            else:
                x += 1
````

#### Grass generation
Replace remaining void tiles with grass tiles.

````python
for y in range(grid_size):
    for x in range(grid_size):
        if map_grid[y][x] == '_':
            map_grid[y][x] = '"'
````

#### Output
Print the map.

````python
for row in map_grid:
    print(''.join(row))
````
